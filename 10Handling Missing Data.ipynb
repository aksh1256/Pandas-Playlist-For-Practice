Handling Missing Data Using Replace and Interpolate

Replace


```python
import pandas as pd 
import numpy as np 
```

```python
var = pd.read_csv("doge.csv")
var
```

```python
# Using thi fill 0.290500 in 0's place
var.replace(to_replace=0, value=0.290500)
```

```python
var
```

```python
var.replace(to_replace="null", value="python")
```

```python
# Using this all serial no replace in 3
# var.replace([sr.no], 3)
```

```python
# Using this all character make a python
var.replace("[A-Za-z]", "python", regex=True)
```

```python
# Using Dictionary in Date column a-z character replace in 22
var.replace({"Date":"[a-z]"},22, regex=True)
```

```python
var
```

```python
# Using this 0 replace using forkword values 10 place value replace 11, 12th place value in Price column
var.replace(0,method="ffill")
```

```python
var
```

```python
# Using this 0 replace using forkword values 13 place value replace 11, 12th place value in Price column
var.replace(0, method="bfill")
```

```python
var
```

```python
var.replace(0,method="ffill", limit=1)
```
