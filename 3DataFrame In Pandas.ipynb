**DataFrame**

```python
# We create DataFrame with the help of List and Dictionary.
# Series are 1D Data Structure in Pandas
# Dataframe are 2D Data Structure in Pandas.
```

```python
import pandas as pd
```

```python
l = [1,2,3,4,5,6,7,8]

var = pd.DataFrame(l)
print(var)
print(type(var))
```

```python
d = {"colors": ['orange','white','black','pink','green','wine','blue'],"popular": [3,1,2,4,5,6,7]}
var1 = pd.DataFrame(d)
print(var1)
```

```python
# When i want to show only color column.
var1 = pd.DataFrame(d,columns=['colors'])
print(var1)
```

```python
d = {"colors": ['orange','white','black','pink','green','wine','blue'],"popular": [3,1,2,4,5,6,7],1:[1,2,3,4,5,6,7]}
var1 = pd.DataFrame(d)
print(var1)
```

```python
print(var1['colors'][2])
```

```python
list = [[1,3,5,7,9],[2,4,6,8,10]]
var2 = pd.DataFrame(list)
print(var2)
print(type(var2))
```

```python
# With the help of Series we create a DataFrame.
sr = {"s":pd.Series([1,2,3,4]),"l":([5,6,7,8])}
var3 = pd.DataFrame(sr)
print(var3)
print(type(var3))
```
