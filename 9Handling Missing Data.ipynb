Handling Missing Values 
(dropna and fillna)

```python
import pandas as pd
import numpy as np
```

```python
var = pd.read_csv("doge.csv")
var
```

```python
# drop null values using dropna() function
var.dropna()
```

```python
# Drop Missing values column using this 
var.dropna(axis=1)
```

```python
# Drop whole empty row using this when any row contains all empty values
var.dropna(how="any")
```

```python
# Using this Price column all null values are drop 
var.dropna(subset=["Price"])
```

```python
# # Using this all null values are drop and create new dataset
var.dropna(inplace=True)
var
```

```python
# When in column one or 2 null values are there then using this drop it
var.dropna(thresh=1)
```

```python
var1 = pd.read_csv("expence.csv")
var1
```

```python
var1.dropna(thresh=2)
var1
```

```python
# Using fillna() we fill values on empty place what we want
var1.fillna("python")
```

```python
var1
```

```python
# When we want to fill anything in column using this what we want 
var1.fillna({"Amount":"java"})
```

```python
# Using this we fill backrow and nextrow data on empty place
var1.fillna(method="ffill")   # bfill
```

```python
# Using this forward columns data on missing values place like as saturday-saturday
var1.fillna(method="ffill", axis=1)
```

```python
var1
```

```python
# Using this forward rows data on missing values place like as 6000-6000
var1.fillna(method="ffill", axis=0)
```

```python
var1
```

```python
# Using this we fill the values on missing place like as 4000-4000
var1.fillna(4000,inplace=True)
var1
```

```python
var2 = pd.read_csv("expence.csv")
var2
```

```python
# Using this we fill values in limited place
var2.fillna("python", limit=1)
```
