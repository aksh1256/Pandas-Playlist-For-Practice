**CSV(Comma Seperated Values) Files In Pandas**

```python
# CSV format is a plain text format in which values are separated by commas(comma seperated values).
```

```python
import pandas as pd
```

```python
dis = {"a":[1,3,5,7,9], "b":[2,4,6,8,10],"c":[1,2,3,4,5], "d":[1,4,9,16,25]}
d = pd.DataFrame(dis)
d
```

```python
# Make CSV File Using
d.to_csv("Test.csv")
```
