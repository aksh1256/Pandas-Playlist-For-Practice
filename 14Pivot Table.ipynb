Melt Function and Pivot Table In Pandas

Melt Function

```python
import pandas as pd
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6], "eng":[10,12,14,15,16,12], "maths":[17,18,19,13,14,16]})
var
```

```python
pd.melt(var)  # Using this we make table vartically.
```

```python
pd.melt(var, id_vars=["days"])
```

```python
pd.melt(var, id_vars=["days"], var_name="Python", value_name="wscubetech")
```

Pivot Table

```python
var1 = pd.DataFrame({"days":[1,2,3,4,5,6], "stu_name":["aksh", "harray", "paddy", "sammy", "tej", "haggy"],
                    "eng":[17,12,14,15,16,12], "maths":[19,18,17,13,14,16]})
var1
```

```python
var1.pivot(index="days", columns="stu_name")
```

```python
var2 = pd.DataFrame({"days":[1,2,3,4,5,6], "stu_name":["aksh", "paddy", "paddy", "aksh", "aksh", "paddy"],
                    "eng":[17,12,14,15,16,12], "maths":[19,18,17,13,14,16]})
var2
```

```python
var2.pivot(index="days", columns="stu_name")
```

```python
var2.pivot(index="days", columns="stu_name", values="eng")
```

```python
var2.pivot(index="days", columns="stu_name", values="maths")
```

```python
var3 = pd.DataFrame({"days":[1,1,1,1,2,2], "stu_name":["aksh", "paddy", "paddy", "aksh", "aksh", "paddy"],
                    "eng":[17,12,14,15,16,12], "maths":[19,18,17,13,14,16]})
var3
```

```python
var3.pivot_table(index="stu_name", columns="days", aggfunc="mean")
```

```python
var3.pivot_table(index="stu_name", columns="days", aggfunc="sum")
```

```python
var3.pivot_table(index="stu_name", columns="days", aggfunc="sum", margins=True)
```

```python
var3.pivot_table(index="stu_name", columns="days", aggfunc="mean", margins=True)
```
