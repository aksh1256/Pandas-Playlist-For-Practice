**Arithmatic Operations In Pandas**

```python
import pandas as pd
```

```python
var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var
```

```python
var["C"] = var["A"] + var["B"]
var
```

```python
var["D"] = var["A"] - var["B"]
var
```

```python
var["E"] = var["A"] * var["B"]
var
```

```python
var["F"] = var["B"] / var["A"]
var
```

```python
var1 = pd.DataFrame({"G":[10,20,30,40],"H":[15,16,17,18]})
var1
```

```python
var1["I"] = var1["G"] <= 20
var1
```

```python
var1["J"] = var1["H"] >= 15
var1
```
