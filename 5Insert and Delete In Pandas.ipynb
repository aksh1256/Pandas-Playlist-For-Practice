**Insert and Delete Functions In Pandas**

**Insert**

```python
import pandas as pd
```

```python
var = pd.DataFrame({"A":[1,2,3,4,5], "B":[9,8,7,6,5]})
var
```

```python
var.insert(2, "C", var["A"])
var
```

```python
var.insert(3, "D", [11,12,13,14,15])
var
```

```python
# Limited amount of Data Copy in var
var1 = pd.DataFrame({"a":[1,3,5,7,9],"b":[2,4,6,8,10]})
var1
```

```python
var1["c"] = var1["a"][:3]
var1
```

**Delete**

```python
# We use pop() Function and Del keyword.
```

```python
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8],"C":[6,7,8,9]})
var2
```

```python
var3 = var2.pop("B")
var3
```

```python
var2
```

```python
del var2["A"]
var2
```
