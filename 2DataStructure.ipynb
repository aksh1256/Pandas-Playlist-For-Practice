```python
import pandas as pd
```

```python
x = [2,3,4,5,6,7,8,9]

var = pd.Series(x)
print(var)
print(type(var))
```

```python
print(var[2])
print(var[3])
```

```python
var = pd.Series(x,index=['a','b','c','d','e','f','g','h'],dtype='float',name='Python')
print(var)
```

```python
d = {"name": ['python', 'c', 'c++', 'java'],"por": [15,14,13,12],"rank": [1,4,3,2]}
var1 = pd.Series(d)
print(var1)
```

```python
s = pd.Series(15,index=[1,2,3,4,5])
print(s)
```

```python
print(s[4])
```

```python
s1 = pd.Series(15,index=[1,2,3,4,5,6,7,8])
s2 = pd.Series(15,index=[1,2,3,4,5])
print(s1+s2)
```
